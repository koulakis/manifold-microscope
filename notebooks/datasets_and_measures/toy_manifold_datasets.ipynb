## Toy datasets
The datasets are loaded as classes which all implement the following methods:

- `normalize_parameters`: Updates the parameters defining the manifold so that it is scaled to have volume 1.
- `sample`: Sample uniform on the d-volume of the manifold.
- `tesselation`: Select samples on the manifold so that they are as spread as possible.
- `measures`: Compute global geometric measures of the manifolds
- `measures_pointwise`: Compute geometric measures on each point of the manifold.
- `dimension`: Get the intrinsic dimension of the manifold.

Additionally, a `plot_manifold` function is provided to sample each manifold, uniformly or with a tesselation and then plot it with colormaps displaying the local values of each geometric measure. Note that there are multiple manifolds and measures which have constant values.

## Circle
- Ambient space: $R^2$
- Manifold dimension: $1$
- Manifold: $S^1$

```python
circle = toy_manifolds.Circle(R=3)
toy_manifolds.plot_manifold(circle)
```

```python
circle = toy_manifolds.Circle(R=3)
toy_manifolds.plot_manifold(circle, sampling_type="uniform")
```

## Moons
- Ambient space: $R^2$
- Manifold dimension: $1$
- Manifold: $S^1\sqcup S^1$

```python
moons = toy_manifolds.Moons(R=1.34)
toy_manifolds.plot_manifold(moons)
```

```python
moons = toy_manifolds.Moons(R=3)
toy_manifolds.plot_manifold(moons, sampling_type="uniform")
```

## Sphere
- Ambient space: $R^3$
- Manifold dimension: $2$
- Manifold: $S^2$

```python
sphere = toy_manifolds.Sphere(R=3)
toy_manifolds.plot_manifold(sphere, N=300)
```

```python
sphere = toy_manifolds.Sphere(R=3)
toy_manifolds.plot_manifold(sphere, N=300, sampling_type="uniform")
```

## Torus
- Ambient space: $R^3$
- Manifold dimension: $2$
- Manifold: $S^1\times S^1$

```python
torus = toy_manifolds.Torus(R=3, r=2.1)
toy_manifolds.plot_manifold(torus, N=500)
```

```python
torus = toy_manifolds.Torus(R=3, r=2.1)
toy_manifolds.plot_manifold(torus, N=500, sampling_type="uniform")
```
